```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

info_df = pd.read_excel('/content/drive/MyDrive/ThermoDataBase/Plantar Thermogram Database.xlsx','DM Group')
TCI_R =info_df.iloc[1:,11].tolist()
TCI_L =info_df.iloc[1:,17].tolist()
dm_idx = info_df.iloc[1:,0].tolist()

```

```python

dm_df = pd.DataFrame({'dm_idx': dm_idx, 'TCI-R': TCI_R, 'TCI-L':TCI_L})
dm_df = dm_df.iloc[:-1]
dm_df
```

```python
cls_R =[]
cls_L =[]
for val in TCI_R:
  if val<=2:
    cls_R.append('1')
  elif 2< val <=3:
    cls_R.append('2')
  elif 3< val <=4:
    cls_R.append('3')
  elif 4< val <=5:
    cls_R.append('4')
  elif val >5:
    cls_R.append('5')

for val in TCI_L:
  if val<=2:
    cls_L.append('1')
  elif 2< val <=3:
    cls_L.append('2')
  elif 3< val <=4:
    cls_L.append('3')
  elif 4< val <=5:
    cls_L.append('4')
  elif val >5:
    cls_L.append('5')
```

```python
dm_df = pd.DataFrame({'dm_idx': dm_idx, 'class-R':cls_R,'class-L':cls_L})
dm_df = dm_df.iloc[:-1]
dm_df
```
